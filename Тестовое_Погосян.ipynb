1) Рекламная кампания стартовала вчера с дневным бюджетом 40 $. Половина бюджета была израсходована к полудню, а 80% оставшегося бюджета было потрачено между полуднем и временем закрытия. Сколько долларов не было потрачено?

    Сколько потрачено к полудню - 20$

    Сколько потрачено между полуднем и временем закрытия - 20*0.8 = 16

    Остаток: 40 - 20 - 16 = 4$


2) 5 идентичных рекламных кампаний работали 24 дня по 6 часов в день, потрачено было 120 долларов. Сколько дней они работали бы на 216 долларов,если бы 9 одинаковых кампаний работали бы по 8 часов в день?

    Один час работы одной компании стоил 120 / (24*6*5) = 1/6 доллара в час

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
import requests
from io import BytesIO
import re
warnings.filterwarnings('ignore')
```

```python
url = 'https://docs.google.com/spreadsheets/d/1t2Bl9Jc0JCWOaeFgADJ2bVdoLdjDTWno/export?format=xlsx'
r = requests.get(url)
df = pd.read_excel(BytesIO(r.content),  engine='openpyxl')
df
```

```python
df = df.drop(columns='Unnamed: 5')
```

```python
df
```

```python
def extract_month_year(status):
    match = re.search(r'([А-Яа-я]+\s\d{4})', status)
    if match:
        return match.group(0)
    return None
```

```python
df['month_year'] = df['status'].apply(extract_month_year)
```

```python
current_month = None
for i in range(len(df)):
    if df.loc[i, 'month_year'] is not None:
        current_month = df.loc[i, 'month_year']
    df.loc[i, 'month_year'] = current_month
```

```python
df = df[df['client_id'].notna()]
```

```python
df.status.unique()
```

```python
df
```

```python
df['receiving_date'] = pd.to_datetime(df['receiving_date'], format='%Y-%m-%d', errors='coerce')
```

```python
# 1. Вычислите общую выручку за июль 2021 по тем сделкам, приход денежных средств которых не просрочен.
df.loc[(df.month_year == 'Июль 2021') & (df.status != 'ПРОСРОЧЕНО')]['sum'].sum().round(2)
```

Общая выручка по непросроченным платежам за июль 2021 года составила 859896.47 рублей

```python
# 2. Как изменялась выручка компании за рассматриваемый период? Проиллюстрируйте графиком.
question_2 = df.loc[(df.status != 'ПРОСРОЧЕНО')].groupby('month_year', as_index=False).agg({'sum':'sum'})
custom_dict = {'Май 2021': 0, 'Июнь 2021': 1, 'Июль 2021': 2, 'Август 2021': 3, 'Сентябрь 2021': 4, 'Октябрь 2021': 5}
question_2 = question_2.sort_values(by='month_year', key=lambda x: x.map(custom_dict))
question_2
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(question_2, x='month_year', y='sum')
plt.title('Изменение выручки за рассматриваемый период \n')
plt.xlabel('Месяц')
plt.ylabel('Выручка')
sns.despine()
```

Видно, что наибольшая выручка была в июле 2021 года, после чего она ее значение стало снижаться. Это может быть связано с сезонностью

```python
# 3. Кто из менеджеров привлек для компании больше всего денежных средств в сентябре 2021?
question_3 = df.loc[(df.status != 'ПРОСРОЧЕНО') & (df.sale != '-') & (df.month_year == 'Июль 2021')]\
                                                                    .groupby('sale', as_index=False)\
                                                                    .agg({'sum' : 'sum'}) \
                                                                    .sort_values('sum', ascending=False)
question_3
```

```python
plt.figure(figsize=(12,5))
sns.barplot(question_3, x='sale', y='sum')
plt.title('Привлеченные менеджерами денежные средства в сентябре 2021 года \n')
plt.xlabel('\n Менеджер')
plt.ylabel('Привлеченные средства \n')
sns.despine()
```

С небольшой разницей Иванов, Смирнов и Петрова заработали привлекли больше всего средств. Меньше всех привлек Соколов - всего 17069.8 рублей

```python
# Заполним отдельную колонку значениями в соответствии с условиями задания
bonuses_list = []
for i in range(len(df)):
        if df.iloc[i]['new/current'] == 'новая' and df.iloc[i]['status'] == 'ОПЛАЧЕНО' and df.iloc[i]['document'] == 'оригинал':
            bonuses_list.append(df.iloc[i]['sum']*0.07)
        elif df.iloc[i]['new/current'] == 'текущая' and df.iloc[i]['status'] != 'ПРОСРОЧЕНО' and df.iloc[i]['document'] == 'оригинал':
            if df.iloc[i]['sum'] > 10000:
                bonuses_list.append(df.iloc[i]['sum'] * 0.05)
            else:
                bonuses_list.append(df.iloc[i]['sum'] * 0.03)
        else:
            bonuses_list.append(0)
df['bonuses'] = bonuses_list
```

```python
df['bonuses'] = bonuses_list
```

```python
df
```

```python
# Я следовал такой логике, что для правильного фильтра необходимо указать сделки, произошедшие в мае и июне, а так же указать, что оригиналы для них на 1 июля еще не пришли. Надеюсь, что все правильно понял)
bonuses_balance = df.loc[(df.month_year.isin(['Май 2021', 'Июнь 2021'])) & (df.receiving_date > '2021-07-01')].groupby('sale', as_index=False) \
                                                                                             .agg({'bonuses' : 'sum'}) \
                                                                                             .sort_values('bonuses', ascending=False)
```

```python
bonuses_balance
```

```python
plt.figure(figsize=(12,5))
ax = sns.barplot(bonuses_balance, x='sale', y='bonuses')
ax.bar_label(ax.containers[0])
plt.title('Остаток каждого из менеджеров на 01.07.2021')
plt.xlabel('\nМенеджер')
plt.ylabel('Остаток')
sns.despine()
```
